### Data Cleaning

```python
cols_to_remove = ['RecipeId', 'AuthorId', 'AuthorName', 'DatePublished', 'Images', 'Keywords']
df = df.drop(columns=cols_to_remove)
df.columns = df.columns.str.strip()
df.head()
```

### Data Transformation

```python
import re
```

```python
for col in ['CookTime', 'PrepTime', 'TotalTime']:
    df[col + "_Minutes"] = df[col].apply(lambda x: 
        (int(re.search(r'PT(\d+)H', x).group(1)) * 60 if re.search(r'PT(\d+)H', x) else 0) +
        (int(re.search(r'(\d+)M', x).group(1)) if re.search(r'(\d+)M', x) else 0)
        if pd.notna(x) else None
    )
```

```python
for col in ['RecipeIngredientQuantities', 'RecipeIngredientParts', 'RecipeInstructions']:
    df[col + "_List"] = df[col].apply(lambda x: x[3:-2].split('", "') if pd.notna(x) and x.startswith('c("') else [x])
```

```python
for col in ['Calories', 'FatContent', 'SaturatedFatContent', 'CholesterolContent', 
            'SodiumContent', 'CarbohydrateContent', 'FiberContent', 'SugarContent', 'ProteinContent']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

### Feature Engineering

#####  Aggregated Nutritional Metrics Per Serving

```python
df['Calories_per_Serving'] = df['Calories'] / df['RecipeServings']
df['Protein_per_Serving'] = df['ProteinContent'] / df['RecipeServings']
df['Fat_per_Serving'] = df['FatContent'] / df['RecipeServings']
df['Carbs_per_Serving'] = df['CarbohydrateContent'] / df['RecipeServings']
```

#####  Macronutrient-to-Calorie Ratios

```python
df['Protein_to_Calorie'] = df['ProteinContent'] / df['Calories']
df['Fat_to_Calorie'] = df['FatContent'] / df['Calories']
df['Carb_to_Calorie'] = df['CarbohydrateContent'] / df['Calories']
```

##### Nutrient Ratios and Densities

```python
df['Protein_Fat_Ratio'] = df['ProteinContent'] / df['FatContent'].replace(0, np.nan)
df['Protein_Carb_Ratio'] = df['ProteinContent'] / df['CarbohydrateContent'].replace(0, np.nan)
df['Fat_Carb_Ratio'] = df['FatContent'] / df['CarbohydrateContent'].replace(0, np.nan)
```

##### Time-Related Features

```python
df['Cook_Prep_Ratio'] = df['CookTime_Minutes'] / df['PrepTime_Minutes'].replace(0, np.nan)
df['Time_Difference'] = df['CookTime_Minutes'] - df['PrepTime_Minutes']
```

##### Ingredient-Based Features

```python
df['Ingredient_Count'] = df['RecipeIngredientParts_List'].apply(lambda x: len(x) if isinstance(x, list) else 0)
df['Unique_Ingredient_Count'] = df['RecipeIngredientParts_List'].apply(
    lambda x: len(set(i.lower().strip() for i in x)) if isinstance(x, list) else 0
)

healthy_ingredients = {
    "kale", "spinach", "broccoli", "carrot", "cauliflower", "bell pepper",
    "cucumber", "tomato", "zucchini", "eggplant", "asparagus", "green beans",
    "brussels sprouts",
    "apple", "banana", "orange", "blueberries", "strawberries", "raspberries",
    "grapes", "kiwi", "mango", "pineapple", "pear", "peach",
    "chicken", "turkey", "tofu", "lentils", "beans", "salmon", "egg",
    "almonds", "walnuts", "peanuts", "quinoa", "edamame"
}
df['Contains_Healthy'] = df['RecipeIngredientParts_List'].apply(
    lambda x: any(ing.lower() in healthy_ingredients for ing in x) if isinstance(x, list) else False
)
```

##### Recipe Complexity

```python
df['Num_Steps'] = df['RecipeInstructions_List'].apply(lambda x: len(x) if isinstance(x, list) else 0)
df['Complexity_Score'] = (df['TotalTime_Minutes'].fillna(0) / 10) + df['Ingredient_Count'] + df['Num_Steps']
```

##### Popularity and User Feedback

```python
df['Popularity_Score'] = df['AggregatedRating'] * np.log(df['ReviewCount'] + 1)
```

##### Interaction Terms

```python
df['Ingredients_Calories_Interaction'] = df['Ingredient_Count'] * df['Calories']
df['Ingredients_Protein_Interaction'] = df['Ingredient_Count'] * df['ProteinContent']
df['Ingredients_Fat_Interaction'] = df['Ingredient_Count'] * df['FatContent']
```
